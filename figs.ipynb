```julia
using Pkg
Pkg.activate("./myenv")
using Revise
using MyPackage
```

```julia
# Define sdf parameters
sdf_type = "debye"
a = 0.3
T = 0.3

# Define TLS parameters
ϵ = 0.0
Δ = 0.2
```

```julia
sdf = "./sdf/$(sdf_type)_a_" * string(a) * "_T_" * string(T) * ".json"
```

```julia
# Define the path to the 'runs' folder
runs_folder = "./runs/eps_" * string(ϵ) * "_Delta_" * string(Δ) * "/$(sdf_type)_a_" * string(a) * "_T_" * string(T) 

# Get all subdirectories (directories only) in the 'runs' folder
dirs = filter(d -> isdir(joinpath(runs_folder, d)), readdir(runs_folder))

# Set the t_max values
t_max_values = [40, 40, 40, 40, 30, 30, 20, 30]

# Loop through each directory and each t_max value
for (i, dir) in enumerate(dirs)
    map_tomo_path = joinpath(runs_folder, dir)
    t_max = t_max_values[i]  # Set t_max based on the index
    
    # Plots
    fig_tomo(map_tomo_path)
    fig_Ks(map_tomo_path, 0, 1; t_max=t_max)
    fig_Us(map_tomo_path; t_max=t_max)
    
    # Checks
    check_tomostates_dynamics(map_tomo_path)
    check_Ks(map_tomo_path, 0, 1; t_max=t_max)
end
```
